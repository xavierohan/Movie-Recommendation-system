# Simple Popularity Based Recommender

```python
# USING IMDBS RATING FORMULA --- RATING = v * R /(v + m) + m * C / (m + v)
# v - No of votes
# m - threshold of No of votes
# R - avg rating
# C - mean of No of votes
m = 434
C = 5.61
def assign_rating(x):
    
    v = metadata[metadata['title'] == x]['vote_count']
    R = metadata[metadata['title'] == x]['vote_average']
    rating = v * R /(v + m) + m * C / (m + v)
    return rating
```

```python
metadata['rating'] = assign_rating(metadata['title']) #calculating ratings for every title
metadata.info()
```

```python
# TOP 15 MOVIES BASED ON USER RATINGS
metadata.sort_values('rating',ascending = False)['title'].head(15)
```

```python
# RECOMMENDING TOP MOVIES BASED ON GENRE OF USERS CHOICE

def genre_recomm(x):
    s = metadata['genres'].str.contains(x)
    t = metadata[s]
    print(t.sort_values('rating',ascending = False)['title'].head(15))
```

```python
metadata = metadata.drop([19730, 29503, 35587]) #illegal ids
df = metadata[['title','genres','id','popularity','rating']]

df = df.sort_values('rating', ascending=False)
df = df.head(10000)  #getiing the best 10000 movies according to rating
```

```python
df.head()
```

```python
#converting 'id' to type int
keywords['id'] = keywords['id'].astype('int')
credits['id'] = credits['id'].astype('int')
df['id'] = df['id'].astype('int')
```

```python
#merging credits and keywords to dataframe df
df = df.merge( credits , on = 'id')
df = df.merge( keywords , on = 'id')
#del df['crew']
df.head()
```

```python
# cleaning up genres,keywords,actors columns
from ast import literal_eval
df['genres'] = df['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df['keywords'] = df['keywords'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df['actors'] = df['cast'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df.head()
```

```python
#cleaning up cast and crew columns
df['cast'] = df['cast'].apply(literal_eval)
df['crew'] = df['crew'].apply(literal_eval)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
df['director'] = df['crew'].apply(get_director)
```

```python
del df['cast']
del df['crew']
```

```python
#converting actor names to lower case and joining first and last names
df['actors'] = df['actors'].apply(lambda x: [str.lower(i.replace(" ", "")) for i in x])
```

```python
#converting director name to lower case and joining first and last names
df['director'] = df["director"].str.replace(" ","")
df['director'] = df['director'].str.lower()
```

```python
df.head()
```

```python
df2 = df.copy()
```

```python
# FUNCTION TO ASSIGN WEIGHT TO DIRECTOR

def director_weight(n):
    df2['director'] = df['director'].apply(lambda x: ( f"{x} " * n ))
director_weight(4)
```

```python
# FUNCTION TO ASSIGN WEIGHT TO GENRES
def genre_weight(n):
    df2['genres'] = df['genres'].apply(lambda x: ( f"{x} " * n ))
genre_weight(3)
```

```python
# FUNCTION TO ASSIGN WEIGHT TO KEYWORDS
def keyword_weight(n):
    df2['keywords'] = df['keywords'].apply(lambda x: ( f"{x} " * n ))
keyword_weight(2)
```

```python
df2.head()
```

```python
#dropping NA values
df2 = df2.dropna(axis=0, subset=['title'])
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
# Creating a new column 'All' 
df2['All'] = df2['genres'].astype(str) + df2['keywords'] .astype(str)+ df2['actors'].astype(str) + df2['director'].astype(str)
```

```python
df2.head()
```

```python
df2['All'] =  df2['All'].apply(lambda x: x.replace('[','').replace(']',','))
df3=df2.copy()
```

```python
#Applying countvectorizer on 'All' to find cosine_similarity between movies
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
count_matrix = count.fit_transform(df2['All'])
```

```python
from sklearn.metrics.pairwise import  cosine_similarity
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
def recommendations(title):
    
    recommended_movies = []
    
    # getting the index of the movie that matches the title
    idx=df2[df2['title'] == title].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar movies
    top_10_indexes = list(score_series.iloc[1:11].index)
    
    # appending the list with the titles of the best 10 matching movies
    for i in top_10_indexes:
        recommended_movies.append(list(df2.index)[i])
        
    return recommended_movies
```
